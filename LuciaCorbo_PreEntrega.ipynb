### 4Ô∏è‚É£ Limpieza de datos

```python
# TODO: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados.
# Sugerencia: aplicar .drop_duplicates(), .str.strip(), .str.replace() seg√∫n el caso.
# Documentar el proceso y los resultados.

# 1) Hago copias para no modificar Dataframes originales
ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()

# 2) Saco filas repetidas
ventas_clean = ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()

# 3) NORMALIZAR
#Texto
def normalizar_texto(df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            #Los caracteres invisibles a 1 espacio
            .str.replace(r"[\u200b\t\r\n]", " ", regex=True)
            # M√°s de un espacios pasa a uno
            .str.replace(" +", " ", regex=True)
            .str.title()
        )
    return df

#Fechas
"""si en la columna esta 'fechas' (aunque tenga algo mas) lo pasa a datetime
como dd/mm/yyyy
"""

def normalizar_fechas(df):
    for col in df.columns:
        if "fecha" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
            df[col] = df[col].dt.strftime("%d/%m/%Y")
    return df

# Precio
"""Normalizo solo precio porque: En Ventas cantidad es float, precio es object.
En Clientes edad es int e ingresos es float. En Marketing costo es float.
El precio object es el √∫nico a normalizar y pasar a float. Primero sacando $ y
despues cambiando "," por "" vacio por si hay miles separados con coma"""
def normalizar_precio(df):
    if "precio" in df.columns:
        df["precio"] = (
            df["precio"]
            .astype(str)
            .str.replace("$", "", regex=False)   # saca el $
            .str.replace(",", "", regex=False)   #cambia 1,000.00 a 1000.00
        )
        df["precio"] = pd.to_numeric(df["precio"], errors="coerce")
    return df


#TODO JUNTO
"""Como uso el bucle for hace que si por ejemplo no hay 'fecha' en alguno de
los Datasets no rompe porque solo no entra al bucle, por eso lo aplico a todos
Igualmente no se si es lo m√°s conveniente en cuanto a optimizaci√≥n"""
def normalizar_todo(df):
    df = normalizar_texto(df)
    df = normalizar_fechas(df)
    df = normalizar_precio(df)
    return df

# 4) Guardar las normalizaciones en los ya limpios
ventas_normalizado   = normalizar_todo(ventas_clean)
clientes_normalizado = normalizar_todo(clientes_clean)
marketing_normalizado = normalizar_todo(marketing_clean)

# 5) Ver resultados

print("*"*25 +" ANTES "+"*"*25+"\n")
analizar_dataframe(ventas, "Ventas")
print("*"*25 +" DESPUES "+"*"*25+"\n")
analizar_dataframe(ventas_normalizado, "Ventas")
print("\n"+"-"*50+"\n")

print("*"*25 +" ANTES "+"*"*25+"\n")
analizar_dataframe(clientes, "Clientes")
print("*"*25 +" DESPUES "+"*"*25+"\n")
analizar_dataframe(clientes_normalizado, "Clientes")
print("\n"+"-"*50+"\n")

print("*"*25 +" ANTES "+"*"*25+"\n")
analizar_dataframe(marketing, "Marketing")
print("*"*25 +" DESPUES "+"*"*25+"\n")
analizar_dataframe(marketing_normalizado, "Marketing")
print("\n"+"-"*50+"\n")


# 6) Reporte (lo que est√° en la pre-entrega resulta para comparar con mis resultados)
def reporte(dfs, nombres):
    resumen = []
    for df, nombre in zip(dfs, nombres):
        filas = len(df)
        columnas = len(df.columns)
        nulos = df.isna().sum().sum()
        duplicados = df.duplicated().sum()

        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    return pd.DataFrame(resumen)

print("*"*25 +" COMPARACI√ìN "+"*"*25+"\n")
print(reporte(
    [ventas, clientes, marketing],
    ["VENTAS Original", "CLIENTES Original", "MARKETING Original"]
))

print(reporte(
    [ventas_clean, clientes_clean, marketing_clean],
    ["VENTAS Copia", "CLIENTES Copia", "MARKETING Copia"]
))
```

### 5Ô∏è‚É£ Transformaci√≥n de datos

```python
# TODO: Aplicar filtros y transformaciones para crear una tabla de ventas
# que muestre solo los productos con alto rendimiento. calcular el percentil 80
# y filtrar los productos que superen ese umbral en ventas.
#quantile(0.8)
# Sugerencia: usar .query() o condiciones con operadores l√≥gicos.

"""Estrategia pre-entrega clase:
Plan paso a paso:
Detectar la columna de producto
(tolerando distintos nombres: producto, id_producto, sku, articulo‚Ä¶).
Calcular ingreso por registro = precio * cantidad.
Agregar por producto para obtener m√©tricas (ingreso_total, unidades,
precio_promedio, registros).
Calcular P80 con quantile(q=0.80).
Filtrar productos con ingreso_total >= P80.
Ordenar de mayor a menor."""


# 1) Detecta columna de producto
def detectar_columna_producto(df):
    posibles = ["producto", "id_producto", "sku", "articulo", "item"]
    cols_l = {c.lower(): c for c in df.columns}
    for c in posibles:
        if c in cols_l:
            return cols_l[c]
    raise ValueError("No se encontro columna de producto, (ni parecido).")

# 2) Hace precio * cant y agrega columna ingreso
def calcular_ingreso(df):
    #df = df.copy()
    df["precio"] = pd.to_numeric(df["precio"], errors="coerce")
    df["cantidad"] = pd.to_numeric(df["cantidad"], errors="coerce")
    df["ingreso"] = df["precio"] * df["cantidad"]
    return df

# 3) Metricas por producto
def resumen_por_producto(df, col_producto):
    return (
        df.groupby(col_producto, dropna=False)
          .agg(ingreso_total=("ingreso", "sum"), #son los ingresos totales del prod
               unidades=("cantidad", "sum"), #todas las cant vendidas
               precio_promedio=("precio", "mean"),
               registros=("ingreso", "size")) #cuantas entradas o filas hay del prod
          .reset_index()
    )

# 4) Filtrar alto rendimiento
def filtrar_alto_rendimiento(tabla, percentil=0.80):
    p80 = tabla["ingreso_total"].quantile(percentil)
    top = (
        tabla.loc[tabla["ingreso_total"] >= p80]
             .sort_values("ingreso_total", ascending=False)
             .reset_index(drop=True)
    )
    return p80, top

# 5) TODO
def tabla_alto_rendimiento(df, percentil=0.80):
    col_producto = detectar_columna_producto(df)
    df_ingreso = calcular_ingreso(df)
    tabla = resumen_por_producto(df_ingreso, col_producto)
    p80, top = filtrar_alto_rendimiento(tabla, percentil)
    return col_producto, p80, tabla, top

# 6) ver resultados
prod_col, p80_ingreso, ventas_resumen, ventas_top = tabla_alto_rendimiento(ventas_normalizado)

print(f"Columna de producto detectada: {prod_col}")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print("‚úÖ Productos de ALTO RENDIMIENTO (top 20% por ingreso):\n")
print("*"*25 +" REULTADO "+"*"*25+"\n")
display(ventas_top.head(20))

```

### 6Ô∏è‚É£ Agregaci√≥n

```python
# TODO: Resumir las ventas por categor√≠a de producto y analizar los ingresos generados.
# Sugerencia: usar .groupby() y .agg() para generar m√©tricas como suma y promedio.

"""Estrategia pre-entrega clase:
Plan paso a paso:
Detectar la columna de categor√≠a.
Asegurar columna ingreso (si no existe, crearla).
groupby(categor√≠a).agg(...) para obtener m√©tricas.
Ordenar por ingreso_total.
Calcular ticket_promedio_por_venta
"""

# 1) Encontrar columna de categor√≠a
def encontrar_columna_categoria(df):
    posibles = ["categoria", "categor√≠a", "rubro", "segmento", "family"]
    m = {c.lower(): c for c in df.columns}
    for c in posibles:
        if c in m:
            return m[c]
    raise ValueError("No se encontr√≥ columna de categor√≠a (ni parecido)")

# 2) Columna ingreso si no existe se crea
def asegurar_ingreso(df):
    #df = df.copy()
    if "ingreso" not in df.columns:
        df["precio"] = pd.to_numeric(df["precio"], errors="coerce")
        df["cantidad"] = pd.to_numeric(df["cantidad"], errors="coerce")
        df["ingreso"] = df["precio"] * df["cantidad"]
    return df


# 3) Resumen por categor√≠a y ticket promedio
def resumen_por_categoria(df):
    cat_col = encontrar_columna_categoria(df)
    df_con_ingreso = asegurar_ingreso(df)

    tabla = (
        df_con_ingreso.groupby(cat_col, dropna=False)
           .agg(
               ingreso_total=("ingreso", "sum"),
               ingreso_promedio=("ingreso", "mean"),
               ingreso_mediana=("ingreso", "median"),
               unidades=("cantidad", "sum"),
               ventas=("ingreso", "size"),
               precio_promedio=("precio", "mean")
           )
           .reset_index()
    )

    # Ticket promedio por venta
    tabla["ticket_promedio_por_venta"] = (
        tabla["ingreso_total"] / tabla["ventas"].replace(0, pd.NA)
    )

    #Con sort ordena desde mayor
    tabla = tabla.sort_values("ingreso_total", ascending=False, na_position="last")
    return cat_col, tabla

# 4) Aplicar a ventas_normalizado y ver resultados
cat_col, resumen_cat = resumen_por_categoria(ventas_normalizado)

print(f"Columna de categor√≠a detectada: {cat_col}")
print("Resumen por categor√≠a:")
display(resumen_cat.head(10))

```
